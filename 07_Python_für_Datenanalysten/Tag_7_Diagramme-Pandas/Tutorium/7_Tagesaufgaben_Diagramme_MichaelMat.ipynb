Aufgabe 1:
Verwende den Datensatz "Data Science Salaries 2024" weiter, lade ihn mit Pandas und speichere das Ergebnis in die Variabel salaries.

```python
salaries = pd.read_csv('data_science_salaries.csv')
```

Aufgabe 2:
Lasse dir das Durchschnittsgehalt je Erfahrungslevel ausgeben (Code vom Vortag),
aber stelle diesmal das Ganze als Balkendiagramm dar.
Ordne dabei die Series neu, sodass diese von Entry-Level zu Executive ansteigt. 
Kümmere dich um Titel und Achsenbeschriftungen!

```python
mean_salary = salaries.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=True)
mean_salary.plot(
    title='XP Level / Mean Salary',
    kind='bar',
    color='green',
    ec='red',
    lw=2,
    xlabel='XP Level',
    ylabel='Mean Salary'
);
```

Aufgabe 3:
Erstelle ein Histogramm der (allgemeinen) Gehaltsverteilung. Achte auf Beschriftungen!
Bonus: Rotiere die x-Achsenbeschriftungen um 45 Grad!

```python
salary_distribution = salaries['salary_in_usd']
salary_distribution.plot(
    title='Salary Distribution',
    kind='hist',
    bins=30,
    ec='black',
    color='blue',
    ylabel='Frequency',
    xlabel='USD',
    rot=45
);
```

Aufgabe 4:
Welchen Anteil am Kuchen aller erfassten Personen machen die verschiedenen Arbeitsmodelle (remote etc.) aus?
Löse in einem ansprechenden Kuchendiagramm und ziehe das kleinste Stück dann aus dem Kuchen heraus!
Beschrifte, rotiere, falls nötig und mache hübsch, falls nicht hübsch genug. ;)

```python
work_models_perc = salaries['work_models'].value_counts()
work_models_perc.plot(
    title='Work Models Percentage',
    kind='pie',
    autopct='%1.1f%%',
    startangle=1,
    explode=[0, 0, 0.3],
    ylabel='Percentage',
    cmap='rainbow_r'
);
```

Aufgabe 5:
Nimm die top 10 Länder mit dem besten Durchschnittsgehalt und stelle sie in einem geeigneten Diagramm dar!

```python
top10_salary_countries = salaries.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False)
top10_countries = top10_salary_countries.head(10)
top10_countries.plot(
    title='TOP 10 Countries by Salaries',
    kind='bar',
    ec='black',
    color='white',
    rot=300
);
```
