## Aufgabe 2
Untersuche den autoscout24-Datensatz mit den neuen Mitteln. Prüfe, ob sich die Gruppenmittelwerte nach Angebotstyp (offerType) signifikant unterscheiden und für welche gilt, dass die Nullhypothese verworfen wird.


```python
cars24 = pd.read_csv('autoscout24.csv')
```

```python
oT_new = cars24[cars24['offerType'] == "New"].describe()
oT_new
```

```python
oT_used = cars24[cars24['offerType'] == "Used"].describe()
oT_used
```

```python
oT_demonstration = cars24[cars24['offerType'] == "Demonstration"].describe()
oT_demonstration
```

```python
oT_pre_registered = cars24[cars24['offerType'] == "Pre-registered"].describe()
oT_pre_registered
```

```python
oT_employees = cars24[cars24['offerType'] == "Employee's car"].describe()
oT_employees
```

```python
# 1. Spezifikation:
# H0 => Es gibt keinen signifikanten Unterschied im Zusammenhang zwischen price und offerType (Gruppen)
# H1 => Es gibt einen signifikanten Unterschied im Zusammenhang zwischen price und offerType (Gruppen)

# 2. Signifikanz-Niveau:
# α = 0.05 (5%)

# 3. Prüfgrößen und Verteilung:
# f-Verteilung
```

```python
# 4. f-Wert & p-Wert:
oT_groups = [cars24['price'][cars24['offerType'] == offerType] for offerType in cars24['offerType'].unique()]
result = st.f_oneway(*oT_groups)
f_wert = result.statistic
p_wert = result.pvalue

print(f"F-Wert => {f_wert}\nP-Wert => {p_wert}")
```

```python
# 5. Entscheidung:
# Es gibt einen signifikanten Unterschied im Zusammenhang zwischen price und offerType
# d.h.
# H0 wird allen Fällen verworfen und H1 angenommen
```

```python
# Post-HOC-Test:
# Tukey-HSD-Test:

comp = mc.MultiComparison(cars24["price"], cars24["offerType"])
post_hoc_res = comp.tukeyhsd()
post_hoc_res.summary()
```
