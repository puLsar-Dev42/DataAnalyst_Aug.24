```python
import pandas as pd
import scipy.stats as st
```

```python
people = pd.read_csv('menschen_masse.csv')
people = people['koerpergroesse_cm']
people.head()
```

```python
# Verteilung der Körpergrößen:
people.hist(bins=100)
# Die Daten sind annähernd normalverteilt.
```

```python
# So zieht man in Pandas zufällige Stichproben aus einem Datensatz:
people.sample(30)
```

```python
# Wir tun so, als wäre people die Gesamtpopulation und unsere samples wären die Stichproben
```

```python
pop_mean = people.mean()
print(pop_mean)
```

```python
means = []

for i in range(10_000):
    sample = people.sample(30)
    sample_mean = sample.mean()
    means.append(sample_mean)
```

```python
means
```

```python
means_series = pd.Series(means)
```

```python
means_series.hist(bins=100)
```

```python
means2 = []

for i in range(10_000):
    sample = people.sample(100)
    sample_mean = sample.mean()
    means2.append(sample_mean)
```

```python
means2_series = pd.Series(means2)
```

```python
means2_series.hist(bins=100)
```
