```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("breast_cancer.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True)
```

```python
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=1)
```

```python
model1 = KNeighborsClassifier()
model1.fit(X_train, y_train)
```

```python
model1.score(X_train, y_train)
```

```python
model1.score(X_test, y_test)
```

```python
predictions = model1.predict(X_test)
```

```python
print(classification_report(y_true=y_test, y_pred=predictions))
```

```python
model2 = KNeighborsClassifier()
```

```python
grid_params = {"n_neighbors": range(1, 30),
               "weights": ["uniform", "distance"],
               "metric": ["euclidean", "manhattan"]}
```

```python
grid_cv = GridSearchCV(
    model2,
    grid_params,
    verbose=1,
    cv=5,
)
```

```python
grid_results = grid_cv.fit(X_train, y_train)
```

```python
grid_results.best_score_
```

```python
grid_results.best_params_
```

```python
model3 = KNeighborsClassifier(n_neighbors=7, metric="manhattan",
                              weights="uniform")
model3.fit(X_train, y_train)
```

```python
model3.score(X_train, y_train)
```

```python
model3.score(X_test, y_test)
```

```python
predictions = model3.predict(X_test)
```

```python
# Fazit: Die Werte haben sich verbessert, nur leider nicht beim recall bösartiger Tumore!
print(classification_report(y_true=y_test, y_pred=predictions))
# Dafür müssten weitere Methoden über Gridsearch angewandt werden!
```
