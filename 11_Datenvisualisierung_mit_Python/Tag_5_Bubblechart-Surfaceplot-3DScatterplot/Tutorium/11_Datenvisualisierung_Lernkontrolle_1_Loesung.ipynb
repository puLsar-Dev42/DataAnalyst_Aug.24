```python
# Deine Imports:
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Datensatz laden:
restaurant = pd.read_csv('restaurant.csv', index_col=0)
restaurant.head()
```

# Aufgabe 1
Bilde folgendes Diagramm nach:
![lineplot_bild](1_lineplot.png)

```python
# Einnahmesummen über die Zeit (je Monat eines Jahres):
ax = sns.lineplot(
    restaurant,
    x='year_month',
    y='total',
    estimator='sum',
    marker='s',
    color='green'
)

ax.set(
    title='Einnahmen aus dem Restaurantbetrieb',
    xlabel=None,
    ylabel='Summe (€)',
)

ax.tick_params('x', labelrotation=45)
```

# Aufgabe 2
Bilde folgendes Diagramm nach:
![barplot_bild](2_barplot.png)

```python
# Summen der Einnahmen je Gericht:
ax = sns.barplot(
    restaurant,
    x='meal_name',
    y='total',
    estimator='sum',
    errorbar=None,
    ec='black',
)

ax.set(
    title='Einnahmen nach Gericht',
    xlabel='Gericht',
    ylabel='Summe (€)',
)

ax.tick_params('x', labelrotation=90)
```

# Aufgabe 3
Bilde folgendes Diagramm nach:
![piechart_bild](3_piechart.png)

```python
# Relative Häufigkeiten der Gerichtarten:
type_ratios = restaurant['type'].value_counts(normalize=True)
type_ratios
```

```python
# Der Plot selbst:
fig, ax = plt.subplots()

ax.pie(
    type_ratios,
    labels=type_ratios.index,
    autopct="%.1f%%",
    explode=[0, 0, 0.5],
    startangle=20,
)

ax.set(title='Percentages of Dishes Sold', )
```

# Aufgabe 4
Bilde folgendes Diagramm nach:
![histplot_bild](4_histplot.png)

```python
ax = sns.histplot(
    restaurant,
    x='tip',
    bins=10,
    hue='type',
    multiple='dodge',
    stat='percent'
)

ax.set(
    title='Trinkgeldhäufigkeiten nach Gerichtart',
    xlabel='Trinkgeld (€)',
    ylabel='Prozent'
);
```

# Aufgabe 5
Bilde folgendes Diagramm nach:
![scatterplot_bild](5_scatterplot.png)

```python
# Scatterplot: Betrag vs Trinkgeld
ax = sns.scatterplot(
    restaurant,
    x='total',
    y='tip',
    hue='type',
    palette='Set1',
    ec='black'
)

ax.set(
    title='Rechnungssummen vs. Trinkgelder',
    xlabel='Preis (€)',
    ylabel='Trinkgeld (€)',
);
```
