```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from fontTools.subset import subset
from matplotlib.pyplot import title
from scipy.stats import alpha
```

```python
dia = sns.load_dataset("diamonds")
dia
```

# Aufgabe 1 
Erstelle das folgende Histogramm!

![histogramm_bild](1_histogramm.png)


```python
dias_by_quality = dia[dia["cut"].isin(["Ideal", "Premium"])]
dias_by_quality = pd.DataFrame(dias_by_quality)
sns.set_style("whitegrid")
plt.figure(figsize=(6, 5))

dia_histo = sns.histplot(
	dias_by_quality,
	x="price",
	hue="cut",
	bins=30,
	kde=True,
	multiple="layer",
	palette=["b", "r"],
	alpha=0.5,
	edgecolor="black"
)

plt.xlabel("Preis")
plt.xlim(0, 17600)
plt.ylabel("Häufigkeit")
plt.ylim(0, 9001)
plt.title("Häufigkeit von Diamanten verschiedener Preisklassen")

plt.legend(
	title="Quality",
	handles=dia_histo.legend_.legend_handles,
	labels=["Ideal", "Premium"]
)

dia_histo;
```

# Aufgabe 2 
Erstelle den folgenden Boxplot!

![boxplot_bild](2_boxplot.png)

```python
plt.figure(figsize=(6.4, 5))
plot4 = sns.boxplot(
	dia,
	y='carat',
	x='cut',
	hue='cut',
	order=["Ideal", "Premium", "Very Good", "Good", "Fair"],
	linewidth=1,
	palette={"Ideal": "#e0d6e6", "Premium": "#c8a5c9", "Very Good": "#d370ad", "Good": "#cc347a", "Fair": "#99164c"},
	flierprops={"marker": "d", "markerfacecolor": "black"},
	fliersize=4,
)

plt.title("Carat von Diamanten nach Qualitätsgrad")
plt.xlabel("Qualität")
plt.ylabel("Carat")

plot4;
```

Erstelle den folgenden Violinplot!

![violinplot_bild](3_violinplot.png)

```python
color_E = dia[dia["color"].isin(["E"])]
```

```python
plt.figure(figsize=(10, 8))
color_E = dia[dia["color"].isin(["E"])]
dia_color_E = sns.violinplot(
	color_E,
	x="color",
	y="price",
	hue="cut",
	palette="crest",
	linewidth=1.5,
	saturation=0.8,
	order=["E"],
	inner='quartile'
)

plt.title("Diamantenpreis der Farbe 'E' nach Qualität aufgetrennt")
plt.xlabel("Farbe")
plt.ylabel("Preis")
plt.legend(title="Qualität");
```
