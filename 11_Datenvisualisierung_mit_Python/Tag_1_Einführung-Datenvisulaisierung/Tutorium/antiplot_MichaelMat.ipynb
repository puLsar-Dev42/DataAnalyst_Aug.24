```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("big_german_cities.csv")
data
```

```python
# Aufgabe 5: Code von Birol!

np.random.seed(42)
years = [str(year) for year in range(1950, 2022, 10)] + ["2024"]
cities = [f"Stadt {i}" for i in range(1, 6)]

data = np.random.rand(len(cities), len(years)) * 100
df = pd.DataFrame(data, columns=years, index=cities)

plt.figure(figsize=(14, 6))

for city in df.index:
	plt.plot(years, df.loc[city], marker='o', linestyle='-', linewidth=3, alpha=0.5)

bar_width = 0.1
x_pos = np.arange(len(years))
bar_data = np.random.rand(len(years)) * 100

plt.bar(x_pos - bar_width / 2, bar_data, width=bar_width, color='r', label="Balkendiagramm", alpha=0.7)

plt.xticks(x_pos, years, rotation=90, fontsize=8)
plt.yticks(fontsize=12)
plt.legend(df.index, loc='center', fontsize=10)
plt.show()
```
