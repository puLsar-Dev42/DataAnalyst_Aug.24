```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import xlabel, ylabel, title
```

```python
# set_style()
sns.set_style('whitegrid')
```

```python
# AUFGABE 1)
df = sns.load_dataset("healthexp")
plt.figure(figsize=(12, 6))

marker = {"Germany": "o", "France": "v", "Great Britain": "s", "Japan": "P", "USA": "*", "Canada": "d"}
sns.lineplot(data=df, x="Year", y="Life_Expectancy", hue="Country", markers=marker, markersize=8, style="Country")

plt.xlabel("Jahr")
plt.ylabel("Lebenserwartung")
plt.title("Lebenserwartung verschiedener LÃ¤nder im Verlauf der Jahre")
plt.legend(title="Land");
```

```python
# AUFGABE 2)
titanic = sns.load_dataset("titanic")

titanic['survived'] = titanic['survived'].map({0: 'No', 1: 'Yes'})
order = ["female" , "male"]
plt.figure(figsize=(8, 6))
sns.countplot(x="sex", hue="survived", data=titanic, palette="muted", ec='black',
			  lw=2, order=order)

plt.title("Titanic passengers divided by sex and if survived")
plt.xlabel("sex")
plt.ylabel("Count")
```

```python
# AUFGABE 3)
dow_jones = sns.load_dataset("dowjones")

# line plot
djay = sns.lineplot(
	dow_jones,
	x="Date",
	y="Price",
	lw=3,
	color="#00FF40",
)

# settings
yticks = np.arange(0, 901, 100)

# labels, title
djay.set(
	xlabel="Jahr",
	ylabel="Preis",
	title="Aktienverlauf des Dow Jones",
	yticks=yticks,
);
```

```python
# AUFGABE 4)
exercise = sns.load_dataset("exercise")

# rel plot
exe = sns.relplot(
	exercise,
	x="time",
	y="pulse",
	col="kind",
	hue="diet",
	kind="line",
)
```
