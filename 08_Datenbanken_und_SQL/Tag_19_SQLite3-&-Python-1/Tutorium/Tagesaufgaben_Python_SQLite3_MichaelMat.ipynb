**c)** Lass dir deine neue SQL-Tabelle mit einer zusätzlichen Spalte anzeigen, welche angibt, ob der Film (ab 120 Minuten Filmlänge) Überlänge hat.

```python
create_table_film_actors = """
		CREATE TABLE IF NOT EXISTS film_actors (
    film_id INT NOT NULL,
    title VARCHAR(50),
    description VARCHAR(100),
    release_year INT NOT NULL,
    language_id INT NOT NULL,
    rental_duration INT NOT NULL,
    rental_rate INT NOT NULL,
    length INT NOT NULL,
    replacement_cost INT NOT NULL,
    rating VARCHAR(10),
    last_update TEXT NOT NULL,
    has_overlength INT
);
"""
cursor.execute(create_table_film_actors)
```

```python
data['last_update'] = pd.to_datetime(data['last_update']).dt.strftime('%Y-%m-%d %H:%M:%S')

for _, row in data.iterrows():
	cursor.execute("""
    INSERT INTO film_actors (film_id, title, description, release_year, language_id, rental_duration, rental_rate, length, replacement_cost, rating, last_update)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
		row['film_id'], row['title'], row['description'], row['release_year'], row['language_id'],
		row['rental_duration'],
		row['rental_rate'], row['length'], row['replacement_cost'], row['rating'], row['last_update']))

cursor.execute("""
UPDATE film_actors
SET has_overlength = (length >= 120);
""")
```

```python
connection.commit()
```

**d)** Erstelle eine neue SQL-Tabelle, in die alle Inhalte der bisherigen Tabelle übernommen werden sowie eine zusätzliche Spalte, welche die durchschnittlichen Wiederbeschaffungskosten (`replacement_cost`) für die verschiedenen Ratings anzeigt.

```python
create_table_film_actor_repl_cost_avg = """
		CREATE TABLE IF NOT EXISTS film_actor_repl_cost_avg AS
SELECT
    film_id,
    title,
    description,
    release_year,
    language_id,
    rental_duration,
    rental_rate,
    length,
    replacement_cost,
    rating,
    last_update,
    (SELECT AVG(replacement_cost) FROM film_actors WHERE rating = fa.rating) AS avg_replacement_cost
FROM film_actors fa;
"""
cursor.execute(create_table_film_actor_repl_cost_avg)
```

```python
cursor.execute('SELECT * FROM film_actor_repl_cost_avg')
```

```python
rows = cursor.fetchall()
for row in rows:
	print(row)
```
